We can then read the AirlinesCluster.csv file and create a Pandas dataframe object. This object will have 6 columns being Balance, BonusMiles, BonusTrans, FlightMiles, and DaysSinceEnroll.

```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['AirlinesCluster.csv']))
```
